```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.figure(figsize=(10, 5))
plt.title("Twitter's Stock Price")
plt.xlabel("Days")
plt.ylabel("Close Price USD ($)")
plt.plot(Twitter["Closing Price"])
plt.show()
```

```python
Twitter = Twitter[["Closing Price"]]
print(Twitter.head(20))
```

```python
futureDays = 15
```

```python
Twitter["Prediction"] = Twitter[["Closing Price"]].shift(-futureDays)
print(Twitter.head())
print(Twitter.tail())
```

```python
import numpy as np
x = np.array(Twitter.drop(["Prediction"], 1))[:-futureDays]
print(x)
```

```python
y = np.array(Twitter["Prediction"])[:-futureDays]
print(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.30)
```

```python
# Creating the decision tree regressor model
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor().fit(xtrain, ytrain)

# creating the Linear Regression model
from sklearn.linear_model import LinearRegression
linear = LinearRegression().fit(xtrain, ytrain)
```

```python
xfuture = Twitter.drop(["Prediction"], 1)[:-futureDays]
xfuture = xfuture.tail(futureDays)
xfuture = np.array(xfuture)
print(xfuture)
```

```python
treePrediction = tree.predict(xfuture)
print("Decision Tree prediction =",treePrediction)
```

```python
linearPrediction = linear.predict(xfuture)
print("Linear regression Prediction =",linearPrediction)
```

```python
predictions = treePrediction
valid = Twitter[x.shape[0]:]
valid["Predictions"] = predictions
plt.figure(figsize=(10, 5))
plt.title("Twitter's Stock Price Prediction Model(Decision Tree Regressor Model)")
plt.xlabel("Days")
plt.ylabel("Close Price USD ($)")
plt.plot(Twitter["Closing Price"])
plt.plot(valid[["Closing Price", "Predictions"]])
plt.legend(["Original", "Valid", "Predictions"])
plt.show()
```

```python
predictions = linearPrediction
valid = Twitter[x.shape[0]:]
valid["Predictions"] = predictions
plt.figure(figsize=(10, 5))
plt.title("Twiiter's Stock Price Prediction Model(Linear Regression Model)")
plt.xlabel("Days")
plt.ylabel("Close Price USD ($)")
plt.plot(Twitter["Closing Price"])
plt.plot(valid[["Closing Price", "Predictions"]])
plt.legend(["Original", "Valid", "Predictions"])
plt.show()
```
